# Lab 3: Prompt Engineering Patterns & Personalization

## Aim
To demonstrate how Large Language Models (LLMs) behave without memory and
how conversation memory allows the model to remember previous prompts
and extend its responses accordingly.

```python
from transformers import pipeline
from langchain_community.llms import HuggingFacePipeline

hf_pipeline = pipeline(
    "text2text-generation",
    model="google/flan-t5-base",
    max_length=128
)

llm = HuggingFacePipeline(pipeline=hf_pipeline)
```

### Step 1: Basic Prompt (No Memory)

```python
prompt_1 = "Explain what a computer is."

print(llm.invoke(prompt_1))
```

### Step 2: Extend Prompt with Persona

```python
prompt_2 = (
    "You are a school teacher.\n"
    "Explain what a computer is."
)

print(llm.invoke(prompt_2))
```

### Step 3: Extend Prompt with Teaching Constraint

```python
prompt_3 = (
    "You are a school teacher.\n"
    "Explain in one short paragraph.\n"
    "Explain what a computer is."
)

print(llm.invoke(prompt_3))
```

### Step 4: Extend Prompt with Cognitive Verifier

```python
prompt_4 = (
    "You are a school teacher.\n"
    "Think step by step before answering.\n"
    "Explain what a computer is."
)

print(llm.invoke(prompt_4))
```

### Step 5: Add Root Prompt Rule

```python
prompt_5 = (
    "Your task is to help with academic learning only.\n"
    "You are a school teacher.\n"
    "Use very simple words.\n"
    "Explain what a computer is."
)

print(llm.invoke(prompt_5))
```

### Step 6: Tested it 


```python
print(llm.invoke(prompt_5.replace(
    "Explain what a computer is.",
    "Explain what the internet is."
)))
```
