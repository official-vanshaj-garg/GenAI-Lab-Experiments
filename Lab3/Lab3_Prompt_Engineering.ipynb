```python
from transformers import pipeline
from langchain_community.llms import HuggingFacePipeline

hf_pipeline = pipeline(
    "text2text-generation",
    model="google/flan-t5-base",
    max_length=256
)

llm = HuggingFacePipeline(pipeline=hf_pipeline)
```

## 1. Persona Prompting

Persona Prompting involves instructing the LLM to assume a specific role
before performing a task. This controls the tone, depth, and expertise
of the response.

```python
from langchain_core.prompts import PromptTemplate

persona_prompt = PromptTemplate(
    input_variables=["topic"],
    template=(
        "You are a senior software engineer with 10 years of experience.\n"
        "Explain {topic} to a second-year engineering student."
    )
)

print(llm.invoke(persona_prompt.format(topic="TCP/IP model")))
```

## 2. Cognitive Verifier Pattern

This pattern forces the LLM to identify sub-questions before answering
a complex query, improving reasoning accuracy.

```python
cognitive_prompt = PromptTemplate(
    input_variables=["question"],
    template=(
        "Before answering the question below, identify the sub-questions "
        "that must be answered to verify it.\n\n"
        "Question: {question}"
    )
)

print(llm.invoke(cognitive_prompt.format(
    question="Is blockchain technology secure?"
)))
```

## 3. Question Refinement Pattern

This pattern asks the LLM to improve a vague prompt to obtain a more
accurate response.

```python
refinement_prompt = PromptTemplate(
    input_variables=["prompt"],
    template=(
        "Suggest a better version of the following question to get a more "
        "accurate and useful answer:\n{prompt}"
    )
)

print(llm.invoke(refinement_prompt.format(
    prompt="Explain machine learning."
)))
```

## 4. Provide New Information and Ask Questions

The model is restricted to the given information and must not use
external knowledge.

```python
info_prompt = PromptTemplate(
    input_variables=["text"],
    template=(
        "Based only on the information below, ask three questions to test "
        "understanding.\n\n{text}"
    )
)

text_data = (
    "Large Language Models are trained on large datasets and generate text "
    "using probability-based predictions. They do not possess true understanding."
)

print(llm.invoke(info_prompt.format(text=text_data)))
```

## 5. Root Prompt

A root prompt defines the primary goal and boundaries of the entire
interaction.

```python
root_prompt = PromptTemplate(
    input_variables=["task"],
    template=(
        "Your primary role is to assist with academic laboratory work on "
        "prompt engineering. Do not provide unrelated information.\n\n"
        "Task: {task}"
    )
)

print(llm.invoke(root_prompt.format(
    task="Explain the importance of persona prompting."
)))
```
