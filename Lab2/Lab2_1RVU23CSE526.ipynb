# Hugging Face Pipeline

### What is Hugging Face Pipeline?
A pipeline is:
"Give text → model understands → gives answer"

### *Loading the transformers library*

```python
from transformers import pipeline

sentiment = pipeline("sentiment-analysis")
```

### *Sentiment Analysis and testing the model*

```python
texts = [
    "I loved the movie!",
    "The product quality was terrible.",
    "The lecture was very informative.",
    "I would not recommend this to anyone."
]

results = sentiment(texts)

for text, result in zip(texts, results):
    print(text)
    print(result)
    print()
```

### *Saving output to CSV file*

```python
import csv

with open("sentiment_results.csv", "w", newline="") as file:
    writer = csv.writer(file)
    writer.writerow(["Text", "Sentiment", "Confidence"])

    for text, result in zip(texts, results):
        writer.writerow([text, result["label"], result["score"]])

print("CSV file created successfully")
```

### *Sentence Embeddings*
Words → numbers
Sentences → vectors
Similar meaning → similar vectors

*Loads a sentence embedding model*

```python
from sentence_transformers import SentenceTransformer, util

model = SentenceTransformer("all-MiniLM-L6-v2")
```

*Creating sentence embeddings*

```python
sentences = [
    "How do I cook pasta?",
    "Best way to make spaghetti",
    "What is the capital of France?"
]
```

*Convert sentences to vectors*

```python
embeddings = model.encode(sentences, convert_to_tensor=True)
print("Embedding shape:", embeddings.shape)
```

*Compare meaning using cosine similarity*

```python
query = "How to cook spaghetti?"
query_embedding = model.encode(query, convert_to_tensor=True)

scores = util.pytorch_cos_sim(query_embedding, embeddings)[0]

for sentence, score in zip(sentences, scores):
    print(f"{score:.3f} -> {sentence}")
```

### *FastAI Text Classification (TRAINING PART)*

*Load dataset*

```python
from datasets import load_dataset
import pandas as pd
from fastai.text.all import *

dataset = load_dataset("imdb")
print(dataset)
```

*Using a small subset (fast execution)*

```python
train_data = dataset["train"].shuffle(seed=42).select(range(4000))
train_df = pd.DataFrame(train_data)
```

*Create DataLoaders*

```python
dls = TextDataLoaders.from_df(
    train_df,
    text_col="text",
    label_col="label",
    valid_pct=0.1,
    bs=32
)
```

```python
from fastai.callback.progress import CSVLogger
```

*Train model*

```python
learn = text_classifier_learner(
    dls,
    AWD_LSTM,
    metrics=accuracy,
    cbs=CSVLogger()
)

# Fix for NBMasterBar error (removes progress bar)
from fastai.callback.progress import ProgressCallback
learn.remove_cbs(ProgressCallback)

learn.fine_tune(1)
```
